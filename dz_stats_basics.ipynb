Задание 2. Первичное изучение данных

Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
df.describe()
```

```python
def remove_outlier(row):
    q1 = df[row].quantile(0.25)
    q3 = df[row].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = df[df[row].between(lower_bound, upper_bound, inclusive=True)].sort_values(row)
    return remove_outliers
```

```python
surgery = remove_outlier('surgery?')
```

Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df.isna().sum()
```

```python
horse_data = pd.read_csv('horse_data.csv', na_values=missing_values, usecols=[0,1,3,4,5,6,10,22], names=['surgery?', 'Age', 'rectal temperature', 'pulse',
                                                                       'respiratory rate', 'temperature of extremities',
                                                                       'pain', 'outcome'])
```

```python
#нет признаков по которым можно догадаться делали ли лошади операцию
#т.к незаполненое значение всего одно, я заменил его на одно из двух возможных. т.е. на больше всего встречающееся
#хотя особой разницы нет, какое из двух значений выбрать
horse_data['surgery?'].fillna(df['surgery?'].mode()[0], inplace=True)
```

```python
# возможно температура может зависить от уровня боли, а также от температуры конечностей
horse_data['rectal temperature'].fillna(df.groupby(['temperature of extremities', 'pain'])['rectal temperature'].transform('median'), inplace=True)
```

```python
#заполнил оставшиеся наны медианными значениями
horse_data['rectal temperature'].fillna(df['rectal temperature'].median(), inplace=True)
```

```python
# исход может зависеть от ректальной температуры и от уровня боли
horse_data['outcome'].fillna(df.groupby(['rectal temperature', 'pain'])['outcome'].transform('median'), inplace=True)
```

```python
# уровень боли зависит от пульса
horse_data['pain'].fillna(df.groupby('pulse')['pain'].transform('median'), inplace=True)
```

```python
# т.к. уровень боли субъективный метод я заполнил оставшиеся наны модальными значениями
horse_data['pain'].fillna(df['pain'].mode()[0], inplace=True)
```

```python
# уровень пульса зависит от боли
horse_data['pulse'].fillna(df.groupby('pain')['pulse'].transform('median'), inplace=True)
```

```python
#заполнил оставшиеся наны медианными значениями
horse_data['pulse'].fillna(df['pulse'].median(), inplace=True)
```

```python
# частота дыхания имеет большие колебания. Определить сложно. Может зависеть от многих факторов. Один из них возраст лошади и возможно пульс
horse_data['respiratory rate'].fillna(df.groupby(['Age', 'pulse'])['respiratory rate'].transform('median'), inplace=True)
```

```python
#заполнил оставшиеся наны медианными значениями
horse_data['respiratory rate'].fillna(df['respiratory rate'].median(), inplace=True)
```

```python
# горячие конечности должны коррелировать с повышенной температурой прямой кишки
horse_data['temperature of extremities'].fillna(df.groupby('rectal temperature')['temperature of extremities'].transform('median'), inplace=True)
```

```python
# т.к. температура конечностей субъективный метод я заполнил оставшиеся наны модальными значениями
horse_data['temperature of extremities'].fillna(df['temperature of extremities'].mode()[0], inplace=True)
```
